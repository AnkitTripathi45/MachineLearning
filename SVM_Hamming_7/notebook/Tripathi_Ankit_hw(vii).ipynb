 ## Question 1.b.(i)
 
 i. Research exact match and hamming score/ loss methods for evaluating multi-
label classification and use them in evaluating the classifiers in this problem.

**The exact match is a loss method where we calculate the proportion of data points or insances where the predicted value is not equal to the actual value. It is a measure of model performance ans it increases when a model correctly predicts the true labels. It extends the idea of accuracy from single label to multilabel classifiers.**

**EMR=1/n∑I(y_true(i)=y_pred(i))**

**In Hamming loss, the number of False Positives and False Negative per instance are computed and then average is taken over the total number of training instances. It takes into account the prediction error as well as the missing error, normalized over the number of classes and training examples.**

**Hamming Loss=1/n∑∑I(y_true[i,j]≠y_pred[i,j])**

###### For Family class

```python
rbf_svm=[]
para = {'C': [10.0**i for i in np.arange(-6, 7, 3)], 'gamma': [i for i in np.arange(0.1, 5, 1)]}
clf = SVC(kernel='rbf', decision_function_shape='ovr', random_state=42)

svm_large1 = GridSearchCV(clf,para,cv=10)
svm_large2 = GridSearchCV(clf,para,cv=10)
svm_large3 = GridSearchCV(clf,para,cv=10)


svm_large1.fit(X_train, Y_train.iloc[:,0])
print(svm_large1.best_params_)
rbf_svm.append(svm_large1.best_params_)

svm_large2.fit(X_train, Y_train.iloc[:,1])
print(svm_large2.best_params_)
rbf_svm.append(svm_large2.best_params_)

svm_large3.fit(X_train, Y_train.iloc[:,2])
print(svm_large3.best_params_)
rbf_svm.append(svm_large3.best_params_)
 
```

```python
ypred_trn=[]
ypred_tst=[]
ans_table = pd.DataFrame(columns=['Label', 'C','Train Hamming Error', 'Test Hamming Error'])
for i in range(0,3) :
    param = {'C': [10.0**i for i in np.arange(np.log10(rbf_svm[i]['C'])-3,np.log10(rbf_svm[i]['C'])+3, 1)], 'gamma': [i for i in np.arange(rbf_svm[i]['gamma']-1,rbf_svm[i]['gamma']+1,0.2)]}
    svm_ = GridSearchCV(clf, param, cv=10)
    svm_.fit(X_train, Y_train.iloc[:,i])
    ypredtrn=svm_.predict(X_train)
    ypredtst=svm_.predict(X_test)
    hl_trn=hamming_loss(Y_train.iloc[:,i],ypredtrn)
    hl_tst=hamming_loss(Y_test.iloc[:,i],ypredtst)
    ypred_trn.append(ypredtrn)
    ypred_tst.append(ypredtst)
    ans_table.loc[len(ans_table)] = [Y_train.iloc[:,i].name, svm_.best_params_['C'], hl_trn, hl_tst]

    
exact_match_ratio_trn=np.all(np.array(ypred_trn).T == np.array(Y_train), axis=1).mean()
exact_match_ratio_tst=np.all(np.array(ypred_tst).T == np.array(Y_test), axis=1).mean()
print('Train Exact Match Ratio: ',exact_match_ratio_trn)
print('Test Exact Match Ratio: ',exact_match_ratio_tst)
```

(b) In each cluster, determine which family is the majority by reading the true labels.

Calculate the average Hamming distance, Hamming score, and Hamming loss5
between the true labels and the labels assigned by clusters.

```python
X_kmeans=df.iloc[:,:22]
Y_kmeans=df.iloc[:,-3:]
```

```python
Final_table=pd.DataFrame(columns=['INumber', 'Best k','Majority for Family','Majority for Genus','Majority for Species', 'Hamming Distance:Family', 'Hamming Distance:Genus', 'Hamming Distance:Species', 'Hamming Score:Family', 'Hamming Score:Genus', 'Hamming Score:Species', 'Hamming Loss:Family', 'Hamming Loss:Genus', 'Hamming Loss:Species'])
```

```python
#Monte Carlo Simulation : performing 50 times
for i in range(1,51) :
    df_kmeans=pd.DataFrame(columns=['k', 'Silhouette Score']) 
    
    #Finding the Optimal K-value
    for k in range(2,52):# Silhouette Score for k=1 is taken as 0
        kmeans= KMeans(n_clusters=k, init='random').fit(X_kmeans)
        pred_kmeans=kmeans.predict(X_kmeans)
        df_kmeans.loc[len(df_kmeans)] = [k, silhouette_score(X_kmeans,pred_kmeans)]
        
    k_opt=int(df_kmeans[df_kmeans['Silhouette Score']==df_kmeans['Silhouette Score'].max()]['k']) 
    
    opt_kmeans = KMeans(n_clusters=k_opt, init='random').fit(X_kmeans)
    
    df_family={}
    for k in np.unique(opt_kmeans.labels_) : #To get unique k value
        temp = X_kmeans[opt_kmeans.labels_==k].index
        df_family[k]=df.loc[temp]['Family'].value_counts().idxmax()
        
    df_genus={}
    for k in np.unique(opt_kmeans.labels_) :
        temp2 = X_kmeans[opt_kmeans.labels_==k].index
        df_genus[k]=df.loc[temp2]['Genus'].value_counts().idxmax()
        
    df_species={}
    for k in np.unique(opt_kmeans.labels_) :
        temp3 = X_kmeans[opt_kmeans.labels_==k].index
        df_species[k]=df.loc[temp3]['Species'].value_counts().idxmax()
        
    ypred_fnl=pd.DataFrame()
    ypred_fnl['Family']=pd.DataFrame(opt_kmeans.labels_, columns=['temp']).replace(df_family)
    ypred_fnl['Genus']=pd.DataFrame(opt_kmeans.labels_, columns=['temp2']).replace(df_genus)
    ypred_fnl['Species']=pd.DataFrame(opt_kmeans.labels_, columns=['temp3']).replace(df_species)
    Final_table.loc[len(Final_table)] = [i, k_opt, list(df_family.values()), list(df_genus.values()), list(df_species.values()), round(hamming(Y_kmeans['Family'], ypred_fnl['Family'])*len(Y_kmeans['Family'])), round(hamming(Y_kmeans['Genus'], ypred_fnl['Genus'])*len(Y_kmeans['Genus'])), round(hamming(Y_kmeans['Species'], ypred_fnl['Species'])*len(Y_kmeans['Species'])), (1-hamming_loss(Y_kmeans['Family'], ypred_fnl['Family'])), (1-hamming_loss(Y_kmeans['Genus'], ypred_fnl['Genus'])), (1-hamming_loss(Y_kmeans['Species'],ypred_fnl['Species'])), hamming_loss(Y_kmeans['Family'], ypred_fnl['Family']), hamming_loss(Y_kmeans['Genus'], ypred_fnl['Genus']), hamming_loss(Y_kmeans['Species'], ypred_fnl['Species'])]
```
